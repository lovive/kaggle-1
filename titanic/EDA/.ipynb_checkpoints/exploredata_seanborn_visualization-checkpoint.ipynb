```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print "load data..."
train = pd.read_csv("../data/train.csv")
train.head(3)
```

### Survived,Pclass,Sex,Embarked

```python
fig = plt.figure()
fig.set(alpha=0.2)
fig.set_size_inches(16, 9)

plt.subplot2grid((2,2),(0,0))
train["Survived"].value_counts().plot(kind='bar')
plt.title('Titanic survived information,1 is survived,0 is unsurvived')
plt.xlabel('survived')
plt.ylabel('numbers')

plt.subplot2grid((2,2),(0,1))
train["Pclass"].value_counts().plot(kind='bar')
plt.title('Titanic pclass information')
plt.xlabel('Pclass')
plt.ylabel('numbers')

plt.subplot2grid((2,2),(1,0))
train["Sex"].value_counts().plot(kind='bar')
plt.title('Titanic Sex information')
plt.xlabel('Sex')

plt.subplot2grid((2,2),(1,1))
train["Embarked"].value_counts().plot(kind='bar')
plt.title('Titanic Embarked information')
plt.xlabel('Embarked')

plt.show()
```

Sex

```python
male = train['Survived'][train["Sex"] == "male"].value_counts()
female = train['Survived'][train["Sex"] == "female"].value_counts()

fig = plt.figure()
fig.set(alpha=0.2)
fig.set_size_inches(16, 9)
#female.plot(kind="bar", stacked=True)
df=pd.DataFrame({'Man':male, 'Female':female})
df.plot(kind="bar", stacked=True)

plt.show()
```
